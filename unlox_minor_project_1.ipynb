**Project 1 : Console-based Smart Calculator**

**Problem Statement** : Create a program which simulates a calculaotr. It should provide a menu with all of the possible operations and accept inputs based on user choice. Additionally, an option to continue  execution should be provided.

**Logic** : Create an infinite while loop and show the menu at the beginning of each iteration. Accept a number as the input and proceed with the respective operation. After operation, ask if the user wants to continue. If the enter "Y", continue with program execution. Else, exit the program.

```python
while True:
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power calculation")
    print("6. Modulus")
    ch = int(input("Enter your choice : "))
    if ch == 1:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of addition : ",x+y)
    elif ch == 2:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of subtraction : ",x-y)
    elif ch == 3:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of multiplication : ",x*y)
    elif ch == 4:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of division : ",x/y)
    elif ch == 5:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of power calculation : ",x**y)
    elif ch == 6:
        x = int(input("Enter the value of 1st number : "))
        y = int(input("Enter the value of 2nd number : "))
        print("Result of modulus calculation : ",x%y)
    else:
        print("Invalid choice")
    cont = input("Continue? (Y to continue, else exit) :")
    if cont not in "yY":
        break
```

**Project 2: Student Marks Analyzer**

**Problem statement** : Write a progam which uses NumPy to analyze student marks data provided by the user.

**Logic** : Accept the marks for a number of students provided by the user, and store it in an array. Convert the array into a NumPy array and perform analysis using NumPy functions sum(), mean(), max() and min().

```python
import numpy as np

def analyze(arr):
    avg_marks = np.mean(arr)
    print("Total marks : ",np.sum(arr))
    print("Average marks : ",avg_marks)
    print("Highest marks : ",np.max(arr))
    print("Lowest marks : ",np.min(arr))

    if avg_marks > 85:
        print("Grade : A")
    elif avg_marks >= 70:
        print("Grade : B")
    elif avg_marks >= 50:
        print("Grade : C")
    else:
        print("Fail")

arr = []
n = int(input("Enter the number of students : "))
for i in range(n):
    marks = int(input(f"Enter the marks for student {i+1}: "))
    arr.append(marks)

analyze(np.array(arr))
```
